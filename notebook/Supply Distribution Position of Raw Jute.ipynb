```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://jutecomm.gov.in/Supply_Distribution_Position_of_Raw_Jute.html"
page = requests.get(url, verify=False) 
soup = BeautifulSoup(page.text, 'html.parser')
print(soup.prettify())
```

We ignored the verification because we didn't install the needed certificates, since it's a public govt website with a valid SSL certification, there's no threat, but if we are handling sensitive info this is not at all recommended.


## Two ways to extract the desired elements from a webpage

### 1. Use "find"

```python
table=soup.find('table', class_ ="table table-bordered table-hover text-center")
table
```

if there are multiple tables in a webpage, inspect the webpage and pick out the class name of ur desired table and add it to the syntax.

### 2. Use "find_all"

```python
column_names=table.find_all("th", {"scope": "col"})
column_names
```

```python
column_names=[title.text.strip().title() for title in column_names]  # added strip to remove whitespace, title to convert it into proper case
column_names
```

```python
column_names.remove("Sl.")
column_names
```

```python
column_data = table.find_all("td")
column_data
```

```python
column_data=[td.get_text(strip=True) for td in column_data]
column_data
```

```python
num_columns=10
structured_column_data=[column_data[i:i + num_columns] for i in range(0, len(column_data), num_columns)]
```

```python
df=pd.DataFrame(structured_column_data, columns=column_names)
df
```
